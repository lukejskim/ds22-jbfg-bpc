### 피보나치 함수
>  
- N번째 피보나치 함수를 구하는 함수
- 피보나치(N)은 바로 전번(N-1)값과 전전번(N-2)값의 더한값

```python
_Figure_("피보나치")
```

```python
def fibonacci(num):
    ret_num = int()
    num = int(num)

    if num == 0:
        ret_num = 0
    elif num == 1:
        ret_num = 1
    else:
        ret_num = fibonacci(num - 1) + fibonacci(num - 2)

    return ret_num
```

```python
for num in range(25):
    
    print("Fibonacci({}) = {}".format(num, fibonacci(num)))
```

### 자료구조 - 스택(Stack)
> 파이썬으로 구현한 스택 자료구조
- 스택은 push와 pop이라는 기능을 가지고 있습니다.
- 스택 자료구조는 말 그대로, 쌓아 올리는 것과 같은 자료구조입니다.
- 즉, push은 item을 쌓아올리는 기능이고, pop은 쌓여져 있는 item에서 제일 위의 것을 꺼내는 작업입니다.
- 따라서 스택 자료구조는 Last In First Out(LIFO) 와 같은 특성을 가지고 있습니다.


```python
_Figure_("스택")
```

##### 스택 구현
- 클래스로 스택(Stack)객체를 만들고, 
- push와 pop기능을 구현하고 
- 해당 스택이 비어있는지 확인하는 isEmpty기능까지 구현합니다.
- 또한 스택이 비어있을때 pop기능을 시도하면 에러메세지를 출력하며 False 값을 리턴하게 하겠습니다.


```python
class Stack:
    def __init__(self):
        self.stack_items = []

    # pop 기능 구현
    def pop(self):
        item_length = len(self.stack_items)

        # 스택이 비어있을때는 에러메세지 출력
        if item_length < 1:
            print("Stack is empty!")
            return False

        # 가장 위에 있는 item을 반환하며 삭제
        result = self.stack_items[item_length - 1]
        del self.stack_items[item_length - 1]
        return result

    # push 기능 구현
    def push(self, x):
        self.stack_items.append(x)

    # isEmpty 기능 구현
    def isEmpty(self):
        return not self.stack_items

```

```python
# 테스트 코드
ds_stack = Stack()
print(ds_stack.stack_items)
ds_stack.push(3)
print(ds_stack.stack_items)
ds_stack.push(5)
print(ds_stack.stack_items)
ds_stack.push(7)
ds_stack.push(9)
ds_stack.push(11)
print(ds_stack.stack_items)
print(ds_stack.isEmpty())
print(ds_stack.pop())
print(ds_stack.pop())
print(ds_stack.pop())
print(ds_stack.pop())
print(ds_stack.pop())
print(ds_stack.pop())
print(ds_stack.isEmpty())
```

### 자료구조 - 큐(Queue)
> 파이썬으로 구현한 큐 자료구조
- 큐 자료구조는 요소(item)을 삽입하는 Enqueue 기능과 요소를 빼내는 Dequeue 기능이 있습니다.
- 처음에 삽입한 요소가 먼저 빠지게 되는 First In First Out(FIFO) 특징을 가지고 있습니다.


```python
_Figure_("큐")
```

##### 큐 구현
- 클래스로 큐(Queue)객체를 만들고, 
- Dequeue 기능과 Dequeue 기능을 포함한, 
- 큐가 비어있는지 확인하는 isEmpty 기능을 구현하였으며 
- 추가로 큐가 비어있을때 Dequeue를 수행하면 에러메세지와 함께 False 값을 리턴


```python
class Queue:
    def __init__(self):
        self.queue_items = []

    # enqueue 기능 구현
    def enqueue(self, x):
        self.queue_items.append(x)
        return None

    # dequeue 기능 구현
    def dequeue(self):
        item_length = len(self.queue_items)
        if item_length < 1:
            print("Queue is empty!")
            return False
        result = self.queue_items[0]
        del self.queue_items[0]
        return result

    # isEmpty 기능 구현
    def isEmpty(self):
        return not self.queue_items
```

```python
ds_queue = Queue()
print(ds_queue.queue_items)
ds_queue.enqueue(1)
print(ds_queue.queue_items)
print(ds_queue.dequeue())
print(ds_queue.queue_items)
ds_queue.enqueue(2)
ds_queue.enqueue(3)
ds_queue.enqueue(4)
ds_queue.enqueue(5)
print(ds_queue.queue_items)
print(ds_queue.dequeue())
print(ds_queue.dequeue())
print(ds_queue.queue_items)
print(ds_queue.dequeue())
print(ds_queue.dequeue())
print(ds_queue.dequeue())
print(ds_queue.isEmpty())

```
