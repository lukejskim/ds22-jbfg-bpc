### <font color='brown'> Boolean indexing </font>
> 불리언 색인

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
names
```

```python
data = randn(7, 4)
data
```

```python
names == 'Bob'
```

```python
data[names == 'Bob']
```

```python
data[names == 'Bob', 2:]
```

```python
data[names == 'Bob', 3]
```

```python
names != 'Bob'
```

```python
# data[-(names == 'Bob')]
data[~(names == 'Bob')]
```

<font color='blue'>
[Tip] NumPy Boolean Negative
<br> '-' 연산자는 사용하지 않으며, ~  혹은 Logical Not 함수를 대신 사용한다.

```python
data[~(names == 'Bob')]
```

```python
data[ (names != 'Bob')]
```

```python
data
```

```python
mask = (names == 'Bob') | (names == 'Will')
mask
```

```python
data[mask]
```

```python
data[data < 0] = 0
data
```

```python
data[names != 'Joe'] = 7
data
```

### <font color='brown'> Fancy indexing </font>
> 팬시 색인
- 팬시 색인은 정수 배열을 사용한 색인을 설명하기 위해 NumPy에서 차용한 단어다.

```python
arr = np.empty((8, 4))
for i in range(8):
    arr[i] = i
arr
```

> 
- 특정한 순서로  로우를 선택하고 싶다면, 그냥 원하는 순서가 명시된 정수가 담긴 ndarray나 리스트를 넘기면 된다.

```python
arr[[4, 3, 0, 6]]
```

```python
arr[[-3, -5, -7]]
```

> 
- 다차원 색인 배열을 넘기는 것은 조금 다르게 동작
- 각각의 색인 튜를에 대응하는 1차원 배열을 선택한 후, reshape 시킨다.

```python
# more on reshape in Chapter 12
arr = np.arange(32).reshape((8, 4))
arr
```

> 
- (1, 0), (5, 3), (7, 1), (2, 2) 에 대응하는 요소 선택

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]]
```

```python
arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]]
```

```python
arr[np.ix_([1, 5, 7, 2], [0, 3, 1, 2])]
```

### <font color='brown'> Transposing arrays and swapping axes </font>
> 배열 전치와 축 바꾸기
- 배열 전치는 데이터를 복사하지 않고 데이터 모양이 바뀐 뷰를 반환하는 특별한 기능
- ndarray는 transpose 메소드와 T라는 이름의 특수한 속성을 가진다.
- 다차원 배열의 경우 transpose 메소드는 튜플로 축 번호를 받아서 치환한다.

```python
arr = np.arange(15).reshape((3, 5))
arr
```

```python
arr.T
```

```python
arr = np.random.randn(6, 3)
arr
```

```python
np.dot(arr.T, arr)
```

```python
arr = np.arange(16).reshape((2, 2, 4))
arr
```

```python
arr.transpose((1, 0, 2))
```

```python
arr
```

```python
arr.swapaxes(1, 2)
```
