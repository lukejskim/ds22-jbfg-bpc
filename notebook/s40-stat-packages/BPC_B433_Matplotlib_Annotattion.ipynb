###  경계 상자 제어

### <font color='#00AA00' size=2>bbox 파라미터 </font>
> <font color='#55AA55' size=2>• facecolor : 상자에 사용하는 컬러 </font>
> <br/><font color='#55AA55' size=2>• edgecolor : 상자 모양의 에지에 사용하는 컬러 </font>
> <br/><font color='#55AA55' size=2>• alpha : 상자와 배경을 혼합하기 위해 투명도를 설정할 때 사용 </font>
> <br/><font color='#55AA55' size=2>• boxstyle : round와 square 중 하나로 상자 스타일을 설정 </font>
> <br/><font color='#55AA55' size=2>• pad : boxstyle을 square로 설정할 때, 텍스트와 상자 옆 사이를 채울 용량을 정의 </font>


```python
# pyplot.text() 함수는 딕셔너리를 입력으로 취하는 bbox 파라미터를 지원한다.
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

box = {
    'facecolor' : '.75',
    'edgecolor' : 'k',
    'boxstyle'  : 'round'
}

plt.text(-0.5, -0.20, 'Backmard minimum', bbox=box)

plt.plot(X, Y, c='k')
plt.show()
```

### <font color='blue'> 화살표 추가

<font color='#00AA00' size=2>arrowprops 파라미터  </font>
<br/><font color='#55AA55' size=2>• arrowstyle : '<-', '<', '-', 'wedge', 'fancy', 'simple'       </font>
<br/><font color='#55AA55' size=2>• facecolor : 화살표에 사용하는 컬러                            </font>
<br/><font color='#55AA55' size=2>• edgecolor : 화살표 모양의 가장자리에 사용하는 컬러            </font>
<br/><font color='#55AA55' size=2>• alpha : 화살표와 배겨을 혼합하기 위해 투명도를 설정할 때 사용 </font>

```python
# pyplot.annotate() : 화살표를 그리는 함수
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.annotate('Backmard minimum',
            ha = 'center',
            va = 'bottom',
            xytext = (-1.5, 3.),
            xy = (0.75, -2.7),
            arrowprops = { 'facecolor' : 'black', 
                           'shrink' : 0.05})

plt.plot(X, Y, c='k')
plt.show()
```

### <font color='blue'> 범례 추가

### <font color='blue'> 선 추가

```python
# 대부분 맷플롯립 플롯팅 함수는 선과 모양 집합을 반환함을 언급
# pyplot.Line2D() : 선을 만들지만, 직접 명시적으로 요청하기까지는 선을 랜더링하지 않음
import matplotlib.pyplot as plt

N = 16
for i in range(N):
    plt.gca().add_line(plt.Line2D((0, i), (N-i, 0), color='.75'))
    
plt.grid(True)
plt.axis('scaled')
plt.show()
```

### <font color='blue'> 모양 추가

```python
# 그림에 모양을 추가하는 방법
import matplotlib.patches as patches
import matplotlib.pyplot as plt

# 원
shape = patches.Circle((0, 0), radius=1., color='.75')
plt.gca().add_patch(shape)

# 직사각형
shape = patches.Rectangle((2.5, -.5), 2., 1., color='.75')
plt.gca().add_patch(shape)

# 타원
shape = patches.Ellipse((0, -2.), 2., 1., angle=45., color='.75')
plt.gca().add_patch(shape)

# 장식상자
shape = patches.FancyBboxPatch((2.5, -2.5), 2., 1., boxstyle='sawtooth', color='.75')
plt.gca().add_patch(shape)

# 모두 표시
plt.grid()
plt.axis('scaled')
plt.show()
```

### <font color='blue'> 다각형으로 작업

```python
# 점 목록으로 정의되는 다각형은 대부분 경로보다 더 복잡
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 8)
points = np.vstack((np.cos(theta), np.sin(theta))).transpose()

plt.gca().add_patch(patches.Polygon(points, color='.75'))

plt.grid(True)
plt.axis('scaled')
plt.show()
```

### <font color='blue'> 경로 속성으로 작업

```python
# 모든 경로는 linewidth, linestyle, edgecolor, facecolor, hatch 등인 여러 속성 가짐
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 6)
points = np.vstack((np.cos(theta), np.sin(theta))).transpose()

plt.gca().add_patch(plt.Circle((0, 0), radius=1., color='.75'))
plt.gca().add_patch(plt.Polygon(points, closed=None, fill=None, lw=3., ls='dashed', edgecolor='k'))

plt.grid(True)
plt.axis('scaled')
plt.show()
```

### <font color='blue'> 눈금 간격 제어

### <font color='blue'> 눈금 레이블 붙이기 제어

```python
# 표준 맷플롯립 눈금 API를 사용해 막대 차트의 눈금을 설정하는 작업
# ticker.Locatoer : 눈금의 위치를 생성
# ticker.Formatter : 눈금용 레이블을 생성
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

name_list  = ('Omar', 'Serguey', 'Max', 'Zhou', 'Abidin')
value_list = np.random.randint(0, 99, size=len(name_list))
pos_list   = np.arange(len(name_list))

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter((name_list)))

plt.bar(pos_list, value_list, color='.75', align='center')
plt.show()
```

### <font color='blue'> 고급 레이블 생성

```python
# 눈금 API의 핵심이 레이블에 관련한 축약어
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

def make_label(value, pos):
    return '%0.1f%%' % (100. * value)
    
ax = plt.axes()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))

X = np.linspace(0, 1, 256)
                             
plt.plot(X, np.exp(-10*X), c='k')
plt.plot(X, np.exp(-5*X), c='k', ls='--')
plt.show()
```

```python
# 눈금 API의 핵심이 레이블에 관련한 축약어
import numpy as np
import datetime
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

start_date = datetime.datetime(2016, 1, 1)

def make_label(value, pos):
    time = start_date + datetime.timedelta(days = 365 * value)
    return time.strftime('%b %y')
    
ax = plt.axes()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))

X = np.linspace(0, 1, 256)
                             
plt.plot(X, np.exp(-10*X), c='k')
plt.plot(X, np.exp(-5*X), c='k', ls='--')

labels = ax.get_xticklabels()
plt.setp(labels, rotation=30.)
plt.show()
```
