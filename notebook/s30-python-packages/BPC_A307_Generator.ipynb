- 리스트 생성

```python
def square_numbers1(nums):
    print('call square_numbers1')
    result = []
    
    for i in nums:
        result.append(i**2)
    
    return result
```

```python
nums = [1, 2, 3, 4, 5]
my_nums1 = square_numbers1(nums)
my_nums1
```

```python
type(my_nums1)
```

```python
sum(my_nums1)
```

- 제너레이터 생성

```python
def square_numbers2(nums):
    print('call square_numbers2')
    for i in nums:
        yield i**2
        
```

```python
nums = [1, 2, 3, 4, 5]
my_nums2 = square_numbers2(nums)
```

```python
type(my_nums2)
```

```python
print(my_nums2)
```

```python
# NameError: name 'my_num2' is not defined
sum(my_num2)
```

```python
# 제너레이터 출력
print(next(my_nums2))
print(next(my_nums2))
print(next(my_nums2))
print(next(my_nums2))
print(next(my_nums2))
```

```python
print(next(my_nums2))
```

- list comprehension 사용 리스트 생성

```python
nums = [1, 2, 3, 4, 5]
```

```python
my_nums3 = [ i for i in nums ]
```

```python
type(my_nums3)
```

```python
for i in my_nums3:
    print(i)
```

- list comprehension 사용 제너레이터 생성

```python
my_nums4 = ( i for i in nums )
```

```python
type(my_nums4)
```

```python
my_nums4
```
