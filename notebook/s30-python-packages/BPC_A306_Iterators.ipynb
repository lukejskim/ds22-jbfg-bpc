#### Iterator Ex2

```python
class IterClass(object):
    
    
    def __init__(self, start, last):
        self.current = start
        self.max = last
        
    # 없으면 'object is not iterable' Exception 발생
    def __iter__(self):       
        return self
    
    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        else:
            self.current += 1
            return self.current-1
        
```

```python
iterClass = IterClass(1, 10)
```

```python
type(iterClass)
```

```python
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
```

```python
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
print(iterClass.__next__())
```
